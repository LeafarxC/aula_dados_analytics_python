```python
import pandas as pd
```

```python
dados = {
    'Cidade' : ['Porto', 'Lisboa', 'Porto', 'Aveiro', 'Lisboa', 'Porto'],
    'Vendas' : [100, 150, 120, 80, 200, 90],
    'Produto' : ['A', 'B', 'A', 'C', 'B', 'A' ]
}

df = pd.DataFrame(dados)
print(df)
```

```python
#Agrupar por Cidades e calcular soma das vendas

vendas_por_cidade = df.groupby('Cidade')['Vendas'].sum()
print(vendas_por_cidade)
```

```python
media_vendas = df.groupby(['Cidade', 'Produto'])['Vendas'].mean()
print(media_vendas)
```
