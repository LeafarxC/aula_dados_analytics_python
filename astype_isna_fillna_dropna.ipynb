```python
import pandas as pd
import numpy as np
```

```python
dados = {
    'Numero' : ['1', '2', '3', '4'], # números em string
    'Preço' : [10.5, 20.0, 15.75, 25.00], # Float
}

df = pd.DataFrame(dados)

print(df)
```

```python
df.info()
```

```python
#converter 'Numero_string'para int

df['Numero'] = df['Numero'].astype(int)

```

```python
print(df.info())
```

```python
#converter preço float para int

df['Preço'] = df['Preço'].astype(int)
```

```python
print(df.info())
```

```python
print(df)
```

```python
#Valores ausentes

df2 = pd.DataFrame({
    'A' : [1, 2, np.nan, 4, 5],
    'B' : [np.nan, 2, 3, 4, np.nan],
    'C' : [1, 2, 3, 4, 5]
})
```

```python
print(df2)
```

```python
#isna()
print(df2.isna())
```

```python
#fillna - preencher com 0 onde tem valores nulos

df_nulos = df2.fillna({'A' : 0})

print(df_nulos)

```

```python
#fillna - preencher com uma média de valores

df_media = df2.fillna(df2.mean())
```

```python
print(df_media)
```

```python
#Remover linhas com valores nan
df_drop_linhas = df2.dropna()
print(df_drop_linhas)
```

```python
#Remover colunas que tenham nan
df_drop_cols = df2.dropna(axis=1)
print(df_drop_cols)
```
