```python
#Pecorrendo um dicionario

pessoas = {
    'Nome' : 'Mário',
    'Idade' : 19,
    'Altura' : 1.75
}
```

```python
print(pessoas)
```

```python
#impressão da chave e dos valores do dicionário

for k in pessoas:
  print(f'{k} : {pessoas[k]}')
```

```python
#Utilizando método Keys()

for chave in pessoas.keys():
  print(chave)
```

```python
#utilizando métodos values()
for valor in pessoas.values():
  print(valor)
```

```python
#utilizando método items()
for chave, valor in pessoas.items():
  print(f'{chave} : {valor}')
```
