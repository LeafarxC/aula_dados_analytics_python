```python
import pandas as pd
```

```python
#criando uma series a partir de uma lista

numeros = [10, 20, 30, 40, 50]

serie1 = pd.Series(numeros) #pd.Series()
```

```python
#Acessar um elemento da serie
print(serie1[3])
```

```python
#criar um Series com indice personalizado

lista_numeros = [100, 200, 300]
index =["a", "b", "c"]

serie_personalizado = pd.Series(lista_numeros, index)
print(serie_personalizado)
```

```python
print(serie_personalizado['a'])
```

```python
#criando uma series a partir de uma dicionario
dados = {'primeiro' : 1, 'segundo' : 2, 'terceiro' : 3}
print(dados)

serie_dict = pd.Series(dados)
```

```python
print(serie_dict)
```

```python
print(serie_dict['segundo'])
```
