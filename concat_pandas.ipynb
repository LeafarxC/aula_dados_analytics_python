```python
import pandas as pd
```

```python
dados1 = {
    'Cidade' : ['Porto', 'Lisboa', 'Porto', 'Aveiro', 'Lisboa', 'Porto'],
    'Vendas' : [100, 150, 120, 80, 200, 90],
    'Produto' : ['Banana', 'Pera', 'Banana', 'Uva', 'Pera', 'Banana']
}
index1 = [0, 1, 2, 3, 4, 5]


dados2 = {
    'Cidade' : ['Braga', 'Coimbra', 'Braga', 'Faro', 'Coimbra', 'Braga'],
    'Vendas' : [120, 170, 130, 90, 220, 110],
    'Produto' : ['Batata', 'Cebola', 'Batata', 'alho', 'Cebola', 'Banana']
}
index2 = [6, 7, 8, 9, 10, 11]
```

```python
df1 = pd.DataFrame(dados1, index1)
df2 = pd.DataFrame(dados2, index2)
```

```python
print(df1)
```

```python
print(df2)
```

```python
concat_linhas = pd.concat([df1, df2])
print(concat_linhas)
```

```python
concat_cols = pd.concat([df1, df2], axis=1)
print(concat_cols)
```
