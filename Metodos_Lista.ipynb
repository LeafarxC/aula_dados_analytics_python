```python
#Tamanho de uma lista
cores = [ 'Azul', 'Amarelo', 'Verde']
print(cores)

print(len(cores))
```

```python
#Inserir dados em uma lista (append)
cores.append('castanho')
print(cores)
```

```python
#Inserir dados em uma posição específica (insert)
frutas = ['laranja', 'abacaxi', 'laranja']
frutas.insert(0, 'pera')
print(frutas)
```

```python
#juntar listas com o extend
cores1 = ['azul', 'amarelo']
cores2 = ['vermelho', 'laranja']

cores1.extend(cores2)

Print(cores1)
```

```python
#Remove dados de uma lista (remove)
frutas = ['laranja', 'pera', 'maçã']

frutas.remove('laranja')

print(frutas)
```

```python
#Remove dados de uma lista (pop)
frutas.pop() #remove o último elemento da lista

frutas.pop(0) #remove o index que for colocado
```

```python
#Remove dados de uma lista (del)
del frutas[0]

del frutas #exclui a lista toda
```

```python
#Limpar todos os itens de uma lista(clear)
frutas.clear()
```
