```python
import pandas as pd
```

```python
dados = {
    'Produto' : ['A', 'B', 'C', 'D'],
    'Preço' : [10, 20, 15, 25],
    'Quantidade_vendida' : [5, 3, 7, 2]
}

df = pd.DataFrame(dados)

print(df)
```

```python
df['Valor_total'] = df['Preço']* df['Quantidade_vendida']
print(df)
```

```python
#Coluna com valor fixo

df['Desconto_aplicado'] = 0.10
print(df)
```

```python
#renomear a coluna produto

df_novo = df.rename(columns={'Produto' : 'Produto_loja'})

print(df_novo)
```

```python
#Modificar o dataframe original usar inplace e não colocar uma nova variável
df.rename(columns={'Produto' : 'Produto_loja'}, inplace=True)

print(df)
```

```python
#Remover Colunas
df = df.drop('Desconto_aplicado', axis=1)
print(df)
```

```python
df.drop(['Quantidade_vendida', 'Valor_total'], axis=1, inplace=True)
print(df)
```
