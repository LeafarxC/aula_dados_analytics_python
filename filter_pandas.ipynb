```python
import pandas as pd
```

```python
dados = {
    'Nomes' : ['Rafael', 'Ana', 'Roberto'],
    'Idade' : [32, 24, 28],
    'Cidade' : ['Faro', 'Lisboa', 'Braga'],
    'Pontuação' : [78, 89, 95]
}

df = pd.DataFrame(dados)
```

```python
print(df)
```

```python
#Filtrar Pessoas com idade maior que 30

filtro_idade = df[df['Idade']>30]
```

```python
print(filtro_idade)
```

```python
#Cidade Lisboa e Pontuação acima de 85
filtro_cid_pont = df[(df['Cidade']== 'Lisboa') & (df['Pontuação'] > 80)]
```

```python
print(filtro_cid_pont)
```

```python
#Cidade Lisboa ou Faro
filtro_cid_cid = df[(df['Cidade']== 'Lisboa') | (df['Cidade'] == 'Faro')]
```

```python
print(filtro_cid_cid)
```

```python
#Comando isin()
filtro_isin = df[df['Cidade'].isin(['Lisboa', 'Faro'])]
print(filtro_isin)
```
