```python
import pandas as pd
```

```python
dados = {
    'Produtos' : ['Maçã', 'Banana', 'Laranja'],
    'Preço' : [10.5, 23.5, 20.0],
    'Stock' : [100, 50, 70]
}

df = pd.DataFrame(dados)
```

```python
print(df)
```

```python
#Ordenar por Preço em ordem crescente

df_preco = df.sort_values(by='Preço')
print(df_preco)
```

```python
#Ordenar por Stock em ordem decrescente
df_qtd = df.sort_values(by='Stock', ascending=False)
print(df_qtd)
```

```python
#Ordenar por múltiplas colunas
df_multi = df.sort_values(by=['Preço', 'Stock'], ascending=[True, False])
print(df_multi)
```
