```python
import pandas as pd
```

```python
#criar um df

dados = {
    'Produtos' : ['A', 'B', 'C'],
    'Preço' : [10.5, 23.5, 20.0],
    'Stock' : [100, 50, 70]
}

index = ['item1', 'item2', 'item3']

df = pd.DataFrame(dados, index)
```

```python
print(df)
```

```python
#Visualizar colunas específicas
#df.Produtos
df['Stock']
```

```python
#Selecionar Multiplas colunas
print(df[['Produtos', 'Stock']])
```

```python
#Comando Loc - Seleção do Item3
print(df.loc['item3'])

```

```python
#Comando Loc para linhas e colunas
print(df.loc[['item1', 'item3'], ['Preço', 'Stock']])
```

```python
#Comando iLoc - para buscar o item3

print(df.iloc[2]) #index
```

```python
#Comando iLoc - para linhas e colunas

print(df.iloc[0:2, 1:3]) #index
```
