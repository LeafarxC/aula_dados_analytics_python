```python
import pandas as pd
```

```python
df_vendas_loja_a = pd.DataFrame({
    "Item" : ['TV', 'Coluna', 'Fogão'],
    "Preço" : [1500, 1200, 1800],
    'Vendas_A' : [100, 150, 200]
}, index = ['TV', 'Coluna', 'Fogão'])
```

```python
df_vendas_loja_b = pd.DataFrame({
    'Vendas_B' : [100, 200, 300],
    'Stock_B' : [20, 40, 50]
}, index = ['TV', 'Coluna', 'M.Lavar'])
```

```python
print(df_vendas_loja_a)
```

```python
print(df_vendas_loja_b)
```

```python
# Left join (Padrão)
left_join = df_vendas_loja_a.join(df_vendas_loja_b, how = 'left')
print(left_join)
```

```python
#Inner join
inner_join = df_vendas_loja_a.join(df_vendas_loja_b, how = 'inner')
print(inner_join)
```

```python
#outer join
outer_join = df_vendas_loja_a.join(df_vendas_loja_b, how = 'outer')
print(outer_join)
```
