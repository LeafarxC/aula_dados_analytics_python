```python
import pandas as pd
```

```python
dados = {
    'Nome': ['Ana', 'Pedro', 'Carlos', 'Daniel'],
    'Idade' : [23, 28, 34, 45],
    'Localidade' : ['Porto', 'Lisboa', 'Coimbra', 'Faro' ]
}
```

```python
print(dados)
```

```python
df = pd.DataFrame(dados)

print(df)
```

```python
#Head(n) - Retorna as primeiras n linhas do df
print(df.head(3))
```

```python
#Tail(n) - Retorna as ultimas n linhas

print(df.tail(2))
```

```python
#shape - Retorna quantidade de linhas e colunas
df.shape
```

```python
#Info - Fornece um resumo do DataFrame

df.info()
```

```python
#Describe() - Retorna estatísticas dobre o DF - Retorna uma estatística descritiva / Mean- Média / std - desvio Padrão

print(df.describe())
```
