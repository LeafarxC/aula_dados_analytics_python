```python
import pandas as pd
```

```python
#criando o nosso primeiro Dataframe a partir de um dicionario
dados = {
    'Nome' : ['Alice', 'Eduardo', 'Carlos', 'Rafael'],
    'Idade' : [25, 30, 19, 27],
    'Cidade' : ['Porto', 'Lisboa', 'Aveiro', 'Braga']
}

print(dados)
```

```python
df = pd.DataFrame(dados) #pd.DataFrame()
```

```python
print(df)
```

```python
#print(df.Nome)
print(df['Cidade'])
```

```python
#criar um DT a partir de uma lista de dicionarios
lista_dict = [
    {'Nome' : 'Rodrigo', 'Idade' : '28', 'Cidade' : 'Porto'},
    {'Nome' : 'Pedro', 'Idade' : '30', 'Cidade' : 'Braga'},
    {'Nome' : 'Rui', 'Idade' : '20', 'Cidade' : 'Aveiro'}
]

df2 = pd.DataFrame(lista_dict)
```
