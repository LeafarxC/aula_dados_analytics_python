```python
#Percorrer uma lista com for
carrro = ['opel', 'tesla', 'Mercedes']
print(carros)

for x in carros:
  print(x)
```

```python
#Percorrer uma lista com range
for x in range(len(carros)):
  print(carros[x])


#Utilizar while
contador = 0
while contador < len(carros):
  print(carros[contador])
  contador += 1
```
