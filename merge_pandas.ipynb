```python
import pandas as pd
```

```python
df_clientes = pd.DataFrame({
    'ID_Clientes' : [1, 2, 3, 4],
    'Nome' : ['João', 'Maria', 'Pedro', 'Ana']
})

df_pedidos = pd.DataFrame({
    'ID_Pedidos' : [101, 102, 103, 104, 105],
    'ID_Clientes' : [1, 2, 3, 5, 2],
    'Produto' : ['Camiseta', 'Calça', 'Sapatilhas', 'Mochila', 'Caderno'],
    'Quantidade' : [2, 1, 3, 2, 4]
})
```

```python
print(df_clientes)
```

```python
print(df_pedidos)
```

```python
#Inner Merge
merge_inner = pd.merge(df_clientes, df_pedidos, on='ID_Clientes', how='inner')
print(merge_inner)
```

```python
#Left Merge
merge_left = pd.merge(df_clientes, df_pedidos, on='ID_Clientes', how='left')
print(merge_left)
```

```python
#Right Merge
merge_right = pd.merge(df_clientes, df_pedidos, on='ID_Clientes', how='right')
print(merge_right)
```

```python
#Outer Merge
merge_outer = pd.merge(df_clientes, df_pedidos, on='ID_Clientes', how='outer')
print(merge_outer)
```
