```python
import pandas as pd
```

```python
#Criar df

dados = {
    'Nome' : ['Carlos', 'Ines', 'Maria', 'Rita'],
    'Idade' : [24, 29, 35, 25],
    'Status' : ['ativo', 'ativo', 'inativo', 'ativo']
}

df = pd.DataFrame(dados)

print(df)

```

```python
#Apply()

def categoria_idade(idade):
  if idade < 30:
    return 'Jovem'
  else:
    return 'Adulto'

df['Categoria_Idade'] = df['Idade'].apply(categoria_idade)
print(df)
```

```python
#map

mapeamento_status = {'ativo' : 'Ativo', 'inativo': 'Inativo'}
df['Novo_Status'] = df['Status'].map(mapeamento_status)
print(df)
```

```python
#Apply - lambda

df['Idade_Dobrada'] = df['Idade'].apply(lambda x : x * 2)
print(df)
```
