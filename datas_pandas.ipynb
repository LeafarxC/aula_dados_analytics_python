```python
import pandas as pd
```

```python
df_datas = pd.DataFrame({
    'Data_string' : ['2025-01-01', '2025-05-05', '2025-01-18'],
    'Evento' : ['Feriado', 'Reunião', 'Lançamento']
})

print(df_datas)
```

```python
#Dtypes

print(df_datas.dtypes)
```

```python
#Converter Data_String para datetime

df_datas['Data'] = pd.to_datetime(df_datas['Data_string'])
print(df_datas)
print(df_datas.dtypes)
```

```python
df_datas_formatos = pd.DataFrame({
    'Data' : ['01/01/2025', '02-jan-2025', '20250118'],
    'valor' : [10,20,30]
})

print(df_datas_formatos)
print(df_datas_formatos.dtypes)
```

```python
df_datas_formatos['Data_Convertida'] = pd.to_datetime(df_datas_formatos['Data'], format='mixed')
print(df_datas_formatos)
```

```python
#Extração de datas
df_partes_data = pd.DataFrame({
    'Timestamp' : pd.to_datetime(['2025-01-15 10:30:00', '2025-01-20 11:30:00', '2025-01-25 12:30:00'])
})

print(df_partes_data)
print(df_partes_data.dtypes)
```

```python
df_partes_data['Ano'] = df_partes_data['Timestamp'].dt.year
df_partes_data['Mes'] = df_partes_data['Timestamp'].dt.month
df_partes_data['Dia'] = df_partes_data['Timestamp'].dt.day
df_partes_data['Hora'] = df_partes_data['Timestamp'].dt.hour
df_partes_data['Dia_semana'] = df_partes_data['Timestamp'].dt.day_of_week
df_partes_data['Nome_dia_semana'] = df_partes_data['Timestamp'].dt.day_name()
print(df_partes_data)
```

```python
df2 = pd.DataFrame({
    'Data' : pd.to_datetime(['2025-01-15', '2025-01-20', '2025-01-25']),
    'valor' : [10,20,30]
})

print(df2)
```

```python
#filtrar por uma data específica

data_filtro = df2[df2['Data'] == '2025-01-15']
print(data_filtro)
```

```python
#Filtrar por um mês específico
data_filtro2 = df2[df2['Data'].dt.month == 1]
print(data_filtro2)
```

```python
#Filtrar por um intervalo de tempo
data_filtro3 = df2[(df2['Data'] >= '2025-01-01') & (df2['Data'] <= '2025-02-15')]
print(data_filtro3)
```

```python
#between()
data_filtro4 = df2[df2['Data'].between('2025-01-01', '2025-02-15')]
print(data_filtro4)
```
